### Introduccion
---
Las variables aleatorias discretas son aquellas que pueden tomar un número finito o contable de valores. Esto significa que entre dos valores consecutivos de la variable no existe un valor intermedio. Por ejemplo, el número de caras al lanzar una moneda o el número de hijos en una familia son variables discretas.

Las variables aleatorias discretas tienen varias propiedades importantes que las caracterizan:

**Función de probabilidad**

$$P(X = x) = p(x)$$

**Función de distribución acumulada**

$$F(x) = P(X \leq x) = \sum_{x_i \leq x} p(x_i)$$

**Esperanza**

$$\mathbb{E}[X] = ∑x_i · p(x_i)$$

**Varianza**

$$Var(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \sum (x_i-\mathbb{E}[X])^2 · p(x_i)$$

Las distribuciones de una variable aleatoria discreta describen cómo se asignan las probabilidades a los posibles valores de la variable. Las principales son:

* Uniforme
* Binomial
* Bernoulli
* Poisson
* Geométrica
* Binomial negativa

## Variable aleatoria Uniforme Discreta
---
Una variable aleatoria $X$ tiene distribución uniforme discreta en el conjunto si su función de masa de probabilidades está dada por
$$f_{X}(k)=\mathbb{P}(X=k)=\begin{cases}
\frac{1}{n} &\quad \text{si } k\in\{x_{1},\dots,x_{n}\} \\
0 &\quad \text{en otro caso}
\end{cases}$$

Tenemos que la notación es la siguiente:

 $$X\sim Unif(x_{1},\dots,x_{n}).$$

Se llama uniforme porque cada uno de los posibles valores de $X$ tienen la misma probabilidad.

Si $X\sim Unif(x_{1},\dots,x_{n})$, en este caso, se tiene:
	$$\mathbb{E}[X] = \sum_{k}kf_{X}(k)=\frac{x_n + x_1}{2}.$$
	y
	$$\text{Var}(X) = \mathbb{E}[(X-\mathbb{E}[X])^{2}] =\sum_{k}(k-\mathbb{E}[X])^{2}f_{X}(k)=\frac{(x_n - x_1+1)^2-1}{12}$$

**Ejemplo**

Imagina que estás organizando un torneo de videojuegos con 8 participantes, y decides asignar aleatoriamente los turnos para jugar. Cada participante tiene el mismo chance de ser seleccionado en cualquier posición (del 1 al 8). Si definimos la variable aleatoria 𝑇 como la posición asignada a un jugador específico, entonces:

**Función de masa de probabilidad:**

$$P(T = k) =
\begin{cases}
\frac{1}{8} & \text{si } k = 1, 2, \dots, 8, \\
0 & \text{en otro caso}.
\end{cases}
$$

**Esperanza matemática:**

$$\mathbb{E}[T] = \frac{1 + 2 + 3 + \dots + 8}{8} = \frac{36}{8} = 4.5.
$$

**Varianza:**

$$\text{Var}(T) = \frac{8^2 - 1}{12} = \frac{63}{12} = 5.25.
$$

**Función de distribución acumulativa:**

La función de distribución acumulativa $𝐹_𝑇$(𝑥) de la variable aleatoria
$𝑇$ (posición asignada) se define como la probabilidad acumulada de que $𝑇$ sea menor o igual a un valor 𝑥. En este caso:
$$F_T(x) = P(T \leq x) =
\begin{cases}
0 & \text{si } x < 1, \\
\frac{\lfloor x \rfloor}{8} & \text{si } 1 \leq x \leq 8, \\
1 & \text{si } x > 8.
\end{cases}
$$

**Probabilidad de que el jugador esté en las primeras 3 posiciones:**

$$P(T \leq 3) = \sum_{k=1}^{3} \frac{1}{8} = \frac{3}{8} = 0.375.
$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Valores posibles de la variable aleatoria
valores = np.arange(1, 9)  # Posiciones del torneo (1 a 8)
probabilidades = np.full(len(valores), 1 / len(valores))  # Probabilidad uniforme

# Función de masa de probabilidad (FDP)
plt.figure(figsize=(10, 5))
plt.bar(valores, probabilidades, color='skyblue', alpha=0.7)
plt.title("Función de Masa de Probabilidad (FDP)", fontsize=14)
plt.xlabel("Valores posibles", fontsize=12)
plt.ylabel("Probabilidad", fontsize=12)
plt.xticks(valores)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Función de distribución acumulativa (FDC)
fdc = np.cumsum(probabilidades)  # Probabilidad acumulada
plt.figure(figsize=(10, 5))
plt.step(valores, fdc, where='post', color='orange', alpha=0.8)
plt.title("Función de Distribución Acumulativa (FDC)", fontsize=14)
plt.xlabel("Valores posibles", fontsize=12)
plt.ylabel("Probabilidad acumulada", fontsize=12)
plt.xticks(valores)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

## Variable aleatoria Binomial Discreta
---
Decimos que una v.a $X$ tiene distribución binomial con párametros $n$ y $p$, si su función de densidad de masa de probabilidades está dada por:
$$\begin{equation*}
	f_{X}(k)=\begin{cases}
	\binom{n}{k}p^{k}(1-p)^{n-k} &\quad \text{si } k=0,1,\dots,n \\
	\qquad 0 &\quad \text{en otro caso}  
	\end{cases}
\end{equation*}$$

donde $n$ es un entero positivo y $0< p< 1$.

La esperanza de $X$ es:
\begin{align*}
\mathbb{E}[X]&= np
\end{align*}

Ahora,
\begin{align*}
\mathbb{E}[X^2]&= (np)^2 +np(1-p),
\end{align*}

por lo que la varianza de $X$ es:
$$\text{Var}(X)=np(1-p).$$

**Ejemplo**

Un dado sesgado se lanza 8 veces, donde la probabilidad de obtener un "6" es 𝑝 = 0.2.
Sea 𝑌 el número de veces que se obtiene un "6". Entonces, 𝑌 tiene una distribución binomial con parámetros 𝑛 = 8 y 𝑝 = 0.2. Por lo tanto:

**Probabilidad de obtener exactamente 2 veces un "6":**
$$ \mathbb{P}(Y = 2) = \binom{8}{2}(0.2)2(0.8)6 = 28 \cdot 0.04 \cdot 0.262144 = 0.2936. $$

**Probabilidad de obtener exactamente 3 veces un "6":**
$$\mathbb{P}(Y = 3) = \binom{8}{3}(0.2)3(0.8)5 = 56 \cdot 0.008 \cdot 0.32768 = 0.1475.$$

**Probabilidad de que 1<𝑌≤4:**
$$𝑃(1<𝑌≤4)=𝑃(𝑌=2)+𝑃(𝑌=3)+𝑃(𝑌=4)$$

Calculamos:
$$\mathbb{P}(Y = 4) = \binom{8}{4}(0.2)4(0.8)4 = 70 \cdot 0.0016 \cdot 0.4096 = 0.0459.$$
Entonces:
$$𝑃(1<𝑌≤4)=0.2936+0.1475+0.0459=0.487$$

**Relación con la función de distribución acumulativa**
$$𝑃(1<𝑌≤4)=𝐹𝑌(4)−𝐹𝑌(1)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

import numpy as np
from scipy.stats import binom

# Parámetros de la distribución binomial
n = 6  # Número de ensayos
p = 0.3  # Probabilidad de éxito

# Cálculo de probabilidades específicas
prob_X_2 = binom.pmf(2, n, p)  # P(X = 2)
prob_X_3 = binom.pmf(3, n, p)  # P(X = 3)

# Cálculo de la probabilidad acumulada para 1 < X ≤ 5
prob_1_X_5 = sum(binom.pmf(k, n, p) for k in range(2, 6))

# Resultados
print(f"P(X = 2): {prob_X_2:.4f}")
print(f"P(X = 3): {prob_X_3:.4f}")
print(f"P(1 < X ≤ 5): {prob_1_X_5:.4f}")

# Relación con la función de distribución acumulativa
F_5 = binom.cdf(5, n, p)  # F(X ≤ 5)
F_1 = binom.cdf(1, n, p)  # F(X ≤ 1)
prob_1_X_5_via_cdf = F_5 - F_1

print(f"P(1 < X ≤ 5) usando FDC: {prob_1_X_5_via_cdf:.4f}")


# Parámetros de la distribución binomial
n = 6  # Número de ensayos
p = 0.3  # Probabilidad de éxito

# Valores posibles de la variable aleatoria
x = np.arange(0, n + 1)

# Función de masa de probabilidad (PMF)
pmf = binom.pmf(x, n, p)

# Función de distribución acumulativa (CDF)
cdf = binom.cdf(x, n, p)

# Gráfica de la PMF
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Función de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('Número de éxitos (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Gráfica de la CDF
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post', color='orange', alpha=0.8, label='CDF')
plt.title('Función de Distribución Acumulativa (CDF)', fontsize=14)
plt.xlabel('Número de éxitos (X)', fontsize=12)
plt.ylabel('Probabilidad acumulada', fontsize=12)
plt.xticks(x)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Variable aleatoria Bernoulli Discreta
---
El modelo probabilístico Bernoulli se aplica a un experimento cuyo espacio muestral está constituido sólo por dos resultados posibles, éxito y fracaso:

Se considerará una v.a $X$ sobre el espacio muestral $$\Omega=\{\text{éxito, fracaso}\}$$
de tal forma que
* $X(\{\text{éxito}\}) = 1$
* $X(\{\text{fracaso}\}) = 0$

Las probabilidades asociadas a este modelo son:

* $\mathbb{P}(\{\text{éxito}\}) = p$
* $\mathbb{P}(\{\text{fracaso}\}) = 1-p$

donde $0<p<1$.


Entonces la función de densidad de masa de probabilidades de $X$ está dada por:

$$f_{X}(k)=\begin{cases}
p &\quad \text{si } k=1 \\
1-p &\quad \text{si } k=0
\end{cases}$$

Tenemos que la notación es la siguiente:

 $$X\sim \text{Ber}(p)$$

 Si $X\sim \text{Ber}(p)$, entonces su esperanza es:
\begin{align*}
\mathbb{E}[X] = \sum_{k=0}^{1}kf_{X}(k) = 1\cdot p + 0\cdot (1-p) = p.
\end{align*}

Y su varianza se calcula usando,
$$\mathbb{E}[X^{2}] = \sum_{k=0}^{1}k^{2}f_{X}(k) = 1^{2}\cdot p + 0^{2}\cdot (1-p) = p.$$
Por lo tanto,
$$\text{Var}(X)=\mathbb{E}[X^{2}]-(\mathbb{E}[X])^{2} = p-p^{2} = p(1-p)$$

**Ejemplo**

Supongamos que lanzamos una moneda no equilibrada, donde la probabilidad de obtener cara es 𝑝 = 0.4 y la probabilidad de obtener cruz es 1 − 𝑝 = 0.6.

Sea 𝑋 una variable aleatoria que toma el valor:
$$X =
\begin{cases}
1 & \text{si cae cara (éxito)}, \\
0 & \text{si cae cruz (fracaso)}.
\end{cases}
$$

Entonces, 𝑋 sigue una distribución de Bernoulli con parámetro 𝑝 = 0.4.

**Esperanza matemática**
$$𝐸[𝑋] = 𝑝 = 0.4.$$

**Varianza:**
$$Var(𝑋) = 𝑝(1−𝑝) = 0.4 ⋅ 0.6 = 0.24$$

La probabilidad de que 𝑋 = 1 (cara) es:
$$𝑃(𝑋=1) = 𝑝 = 0.4.$$

La probabilidad de que 𝑋 = 0 (cruz) es:
$$𝑃(𝑋=0)=1−𝑝=0.6$$

```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Parámetro de la distribución Bernoulli
p = 0.4  # Probabilidad de éxito (cara)

# Valores posibles de la variable aleatoria
x = [0, 1]  # 0: fracaso, 1: éxito

# Función de masa de probabilidad (PMF)
pmf = bernoulli.pmf(x, p)

# Función de distribución acumulativa (CDF)
cdf = bernoulli.cdf(x, p)

# Mostrar resultados
print(f"PMF (P(X=0)): {pmf[0]:.4f}")
print(f"PMF (P(X=1)): {pmf[1]:.4f}")
print(f"CDF (P(X<=0)): {cdf[0]:.4f}")
print(f"CDF (P(X<=1)): {cdf[1]:.4f}")

# Gráfica de la PMF
plt.figure(figsize=(8, 4))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Función de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('Valores posibles (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Gráfica de la CDF
plt.figure(figsize=(8, 4))
plt.step(x, cdf, where='post', color='orange', alpha=0.8, label='CDF')
plt.title('Función de Distribución Acumulativa (CDF)', fontsize=14)
plt.xlabel('Valores posibles (X)', fontsize=12)
plt.ylabel('Probabilidad acumulada', fontsize=12)
plt.xticks(x)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Variable aleatoria Poisson Discreta
---
Esta variable aleatoria toma los valores sobre el conjunto $\{0,1,2,\dots\}$ y tiene un parámetro $\lambda>0$, el cual representa el número de veces que se espera que ocurra un evento durante un intervalo dado.

La función de masa de probabilidades de una variable aleatoria Poisson, $X$, está dada por:
$$f_{X}(k)=\begin{cases}
	e^{-\lambda}\frac{\lambda^{k}}{k!} &\quad \text{si } k=0,1,2,\dots \\
	0 &\quad \text{en otro caso}  
\end{cases}$$

En efecto, es una función de masa de probabilidades debido a que
$$\begin{align*}
\sum_{k=0}^{\infty}e^{-\lambda}\frac{\lambda^{k}}{k!} &= e^{-\lambda}\sum_{k=0}^{\infty}\frac{\lambda^{k}}{k!} \\ &= e^{-\lambda}\underbrace{\left(1+\lambda+\frac{\lambda^{2}}{2!}+\frac{\lambda^{3}}{3!}+\cdots\right)}_{\text{desarrollo de Taylor de }e^{\lambda}} \\ &= e^{-\lambda}e^{\lambda}=1.
\end{align*}$$

**Ejemplo**

Un centro de llamadas recibe, en promedio, 5 llamadas por minuto. Queremos calcular la probabilidad de que el centro reciba exactamente 3 llamadas en un minuto.

*Fórmula de la distribución de Poisson:*

La probabilidad de que ocurran 𝑘 eventos en un intervalo dado está dada por:

$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$
donde:

𝜆 es el número promedio de eventos (en este caso, 5 llamadas por minuto),

𝑘 es el número de eventos que queremos calcular (en este caso, 3 llamadas),

𝑒 es la base del logaritmo natural (≈2.718).

Sustituyendo los valores:
$$P(X = 3) = \frac{5^3 e^{-5}}{3!} = \frac{125 \cdot e^{-5}}{6}
$$


$$P(X = 3) \approx \frac{125 \cdot 0.0067}{6} \approx 0.1404
$$

**Esperanza matemática y varianza:**

Ambas son iguales a 𝜆. En este caso,
$$𝐸[𝑋]=Var(𝑋)=5$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetro de la distribución Poisson
lambda_poisson = 5  # Promedio de llamadas por minuto

# Valores posibles de la variable aleatoria
x = np.arange(0, 15)  # Rango de valores para X (0 a 14)

# Función de masa de probabilidad (PMF)
pmf = poisson.pmf(x, lambda_poisson)

# Mostrar resultados para un valor específico
k = 3  # Número de llamadas
prob_k = poisson.pmf(k, lambda_poisson)
print(f"P(X = {k}): {prob_k:.4f}")

# Gráfica de la PMF
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Función de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('Número de llamadas (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Variable aleatoria Geométrica Discreta
---
El modelo probabilístico geométrico se aplica a un experimento que consiste en una secuencia de ensayos independientes, donde cada ensayo tiene dos posibles resultados: éxito y fracaso.

Se considerará una variable aleatoria 𝑋 sobre el espacio muestral:
$$\Omega = \{ \text{primer éxito en el ensayo } k \mid k \in \mathbb{N} \}
$$
De tal forma que:
$$X = k \quad \text{si el primer éxito ocurre en el ensayo } k.
$$
Las probabilidades asociadas a este modelo son:
$$P(X = k) = p(1-p)^{k-1}, \quad \text{para } k = 1, 2, 3, \dots
$$
donde
$0<𝑝<1$

Entonces, la función de masa de probabilidad de 𝑋 está dada por:
$$f_X(k) =
\begin{cases}
p(1-p)^{k-1} & \text{si } k \geq 1, \\
0 & \text{en otro caso}.
\end{cases}
$$

Tenemos que la notación es la siguiente:
$$𝑋∼Geo(𝑝)$$

**Esperanza matemática:**
$$\mathbb{E}[X] = \sum_{k=1}^{\infty} k \cdot f_X(k) = \sum_{k=1}^{\infty} k \cdot p(1-p)^{k-1}.
$$

Usando propiedades de series geométricas, se obtiene:
$$\mathbb{E}[X] = \frac{1}{p}.$$

**Varianza:**
$$\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2.$$

Donde:

$$\mathbb{E}[X^2] = \sum_{k=1}^{\infty} k^2 \cdot f_X(k).$$

Usando propiedades de series geométricas, se obtiene:
$$\text{Var}(X) = \frac{1-p}{p^2}.$$

**Ejemplo**

Un estudiante está intentando pasar un examen que tiene una probabilidad de éxito del 𝑝 = 0.3 en cada intento. Queremos modelar el número de intentos necesarios hasta que el estudiante pase el examen por primera vez.

**Función de masa de probabilidad:**

La probabilidad de que el estudiante pase el examen en el 𝑘-ésimo intento está dada por:
$$P(X = k) = p(1-p)^{k-1}, \quad \text{para } k = 1, 2, 3, \dots$$

**Probabilidad de pasar en el primer intento (𝑘 = 1):**
$$P(X = 1) = 0.3(1-0.3)^{1-1} = 0.3$$

**Probabilidad de pasar en el segundo intento (𝑘 = 2):**
$$P(X = 2) = 0.3(1-0.3)^{2-1} = 0.3 \cdot 0.7 = 0.21$$

**Probabilidad de pasar en el tercer intento (𝑘 = 3):**
$$P(X = 3) = 0.3(1-0.3)^{3-1} = 0.3 \cdot 0.7^2 = 0.3 \cdot 0.49 = 0.147$$

**Esperanza matemática:**
$$\mathbb{E}[X] = \frac{1}{p} = \frac{1}{0.3} \approx 3.33$$

**Varianza:**
$$\text{Var}(X) = \frac{1-p}{p^2} = \frac{1-0.3}{0.3^2} = \frac{0.7}{0.09} \approx 7.78$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parámetro de la distribución geométrica
p = 0.3  # Probabilidad de éxito

# Valores posibles de la variable aleatoria
x = np.arange(1, 15)  # Rango de valores para X (1 a 14)

# Función de masa de probabilidad (PMF)
pmf = geom.pmf(x, p)

# Función de distribución acumulativa (CDF)
cdf = geom.cdf(x, p)

# Mostrar resultados para un valor específico
k = 3  # Número de intentos hasta el primer éxito
prob_k = geom.pmf(k, p)
print(f"P(X = {k}): {prob_k:.4f}")

# Gráfica de la PMF
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Función de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('Número de intentos (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Gráfica de la CDF
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post', color='orange', alpha=0.8, label='CDF')
plt.title('Función de Distribución Acumulativa (CDF)', fontsize=14)
plt.xlabel('Número de intentos (X)', fontsize=12)
plt.ylabel('Probabilidad acumulada', fontsize=12)
plt.xticks(x)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Variable aleatoria Binomial Negativa Discreta
---
El modelo probabilístico de la binomial negativa se aplica a un experimento que consiste en una secuencia de ensayos independientes, donde cada ensayo tiene dos posibles resultados: éxito y fracaso. Este modelo se utiliza para determinar el número de fracasos antes de alcanzar un número fijo de éxitos.

Se considerará una variable aleatoria $𝑋$ sobre el espacio muestral:
$$\Omega = \{\text{número de fracasos antes de alcanzar } r \text{ éxitos}\}$$

De tal forma que:
$$X = k \quad \text{si se producen } k \text{ fracasos antes de alcanzar } r \text{ éxitos}.$$

Las probabilidades asociadas a este modelo son:
$$P(X = k) = \binom{k+r-1}{r-1} p^r (1-p)^k, \quad \text{para } k = 0, 1, 2, \dots$$


donde:

𝑟 es el número fijo de éxitos,

𝑝 es la probabilidad de éxito en cada ensayo,

𝑘 es el número de fracasos antes de alcanzar 𝑟 éxitos,

$\binom{k+r-1}{r-1}$ es el coeficiente binomial.

Tenemos que la notación es la siguiente:
$$𝑋∼NB(𝑟,𝑝)$$

**Esperanza matemática:**
$$\mathbb{E}[X] = \frac{r(1-p)}{p}.$$

**Varianza:**
$$\text{Var}(X) = \frac{r(1-p)}{p^2}.$$

**Ejemplo**

Un vendedor tiene una probabilidad de éxito del 𝑝 = 0.25 en cada llamada telefónica. Queremos calcular la probabilidad de que el vendedor realice 𝑘 = 4 llamadas fallidas antes de lograr 𝑟 = 3 ventas exitosas.

La probabilidad de que el vendedor realice 𝑘 = 4 llamadas fallidas está dada por:

$$P(X = 4) = \binom{4+3-1}{3-1} (0.25)^3 (1-0.25)^4$$

Sustituyendo los valores:

$$P(X = 4) = \binom{6}{2} (0.25)^3 (0.75)^4$$

Calculamos el coeficiente binomial:

$$\binom{6}{2} = \frac{6!}{2!(6-2)!} = \frac{6 \cdot 5}{2 \cdot 1} = 15$$

Calculamos las potencias:

$$(0.25)^3 = 0.015625, \quad (0.75)^4 = 0.31640625$$

Sustituimos todo:

$$P(X = 4) = 15 \cdot 0.015625 \cdot 0.31640625 = 0.0741$$

**Esperanza matemática:**
$$\mathbb{E}[X] = \frac{r(1-p)}{p} = \frac{3(1-0.25)}{0.25} = \frac{3 \cdot 0.75}{0.25} = 9$$

**Varianza:**
$$\text{Var}(X) = \frac{r(1-p)}{p^2} = \frac{3(1-0.25)}{0.25^2} = \frac{3 \cdot 0.75}{0.0625} = 36$$

```python
import numpy as np
from scipy.stats import nbinom
import matplotlib.pyplot as plt

# Parámetros de la distribución binomial negativa
r = 3  # Número de éxitos
p = 0.25  # Probabilidad de éxito

# Valores posibles de la variable aleatoria
x = np.arange(0, 20)  # Número de fracasos (k)

# Función de masa de probabilidad (PMF)
pmf = nbinom.pmf(x, r, p)

# Función de distribución acumulativa (CDF)
cdf = nbinom.cdf(x, r, p)

# Mostrar resultados para un valor específico
k = 4  # Número de fracasos
prob_k = nbinom.pmf(k, r, p)
print(f"P(X = {k}): {prob_k:.4f}")

# Gráfica de la PMF
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Función de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('Número de fracasos (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Gráfica de la CDF
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post', color='orange', alpha=0.8, label='CDF')
plt.title('Función de Distribución Acumulativa (CDF)', fontsize=14)
plt.xlabel('Número de fracasos (X)', fontsize=12)
plt.ylabel('Probabilidad acumulada', fontsize=12)
plt.xticks(x)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

# Variales aleatorias continuas
---

Una variable aleatoria continua es aquella que puede tomar cualquier valor en un intervalo real.

Las variables aleatorias continuas tienen varias propiedades importantes que las caracterizan:

**Función de densidad:**
$$f(x) \geq 0 , con  \int f(x)\,dx = 1$$

**Función de distribución acumulada:**  
 $$ F(x) = P(X \leq x) = \int_{-\infty}^x f(t)\,dt $$

 **Esperanza:**  
  $$ \mathbb{E}[X] = \int x \cdot f(x)\,dx $$

  **Varianza:**  
  $$ \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \int (x - \mathbb{E}[X])^2 \cdot f(x)\,dx $$

  Las distribuciones de una variable aleatoria continua describen cómo se asignan las densidades de probabilidad a los posibles valores de la variable. En lugar de trabajar con probabilidades específicas para valores individuales (como en el caso discreto), las distribuciones continuas asignan densidades a intervalos de valores.

  Las principales son:
  
- Uniforme  
- Normal  
- Exponencial  
- Gamma  
- Beta

## Variable aleatoria Uniforme Continua
---
La distribución uniforme continua describe una variable aleatoria 𝑋 que toma valores en un intervalo $[𝑎,𝑏]$ con igual densidad de probabilidad en todo el intervalo. Esto significa que cualquier subintervalo de igual longitud tiene la misma probabilidad.

**Función de densidad de probabilidad (FDP)**

La función de densidad de probabilidad está definida como:
$$f(x) =
\begin{cases}
\frac{1}{b-a}, & \text{si } a \leq x \leq b, \\
0, & \text{en otro caso}.
\end{cases}
$$

**Probabilidad en un intervalo:**

La probabilidad de que 𝑋 tome un valor en un intervalo $[𝑐,𝑑]$ (donde $𝑎≤𝑐≤𝑑≤𝑏$) es:

$$P(c \leq X \leq d) = \int_{c}^{d} f(x) \, dx = \int_{c}^{d} \frac{1}{b-a} \, dx = \frac{d-c}{b-a}$$

**Esperanza matemática:**

La esperanza matemática de $𝑋$ es:
$$\mathbb{E}[X] = \frac{a+b}{2}$$

**Varianza:**

La varianza de $𝑋$ es:
$$\text{Var}(X) = \frac{(b-a)^2}{12}$$

**Ejemplo**

Supongamos que un cliente llega a una tienda y tiene que esperar un tiempo antes de ser atendido. El tiempo de espera sigue una distribución uniforme continua entre 𝑎 = 0 y 𝑏 = 10 minutos. Queremos calcular:

- La probabilidad de que el cliente espere entre 3 y 7 minutos.
- La esperanza matemática y la varianza del tiempo de espera.

La probabilidad de que el cliente espere entre 3 y 7 minutos se calcula como:
$$P(3 \leq X \leq 7) = \frac{7-3}{10-0} = \frac{4}{10} = 0.4$$

Esto significa que hay un 40% de probabilidad de que el cliente espere entre 3 y 7 minutos.

**Esperanza matemática:**

La esperanza matemática del tiempo de espera es:
$$\mathbb{E}[X] = \frac{a+b}{2} = \frac{0+10}{2} = 5$$

En promedio, el cliente esperará 5 minutos.

**Varianza:**

La varianza del tiempo de espera es:
$$\text{Var}(X) = \frac{(b-a)^2}{12} = \frac{(10-0)^2}{12} = \frac{100}{12} \approx 8.33$$



```python
# Parámetros de la distribución uniforme continua
a = 0  # Límite inferior
b = 10  # Límite superior

# Cálculo de la probabilidad en el intervalo [3, 7]
c = 3
d = 7
probabilidad = (d - c) / (b - a)
print(f"P(3 ≤ X ≤ 7): {probabilidad:.2f}")

# Cálculo de la esperanza matemática
esperanza = (a + b) / 2
print(f"Esperanza matemática (E[X]): {esperanza:.2f}")

# Cálculo de la varianza
varianza = ((b - a) ** 2) / 12
print(f"Varianza (Var[X]): {varianza:.2f}")

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución uniforme continua
a = 0  # Límite inferior
b = 10  # Límite superior

# Crear valores de x dentro del intervalo [a, b]
x = np.linspace(a, b, 1000)
f_x = np.full_like(x, 1 / (b - a))  # FDP constante

# Graficar la función de densidad de probabilidad
plt.figure(figsize=(8, 5))
plt.plot(x, f_x, color='blue', label='FDP: $f(x) = \\frac{1}{b-a}$')
plt.fill_between(x, f_x, color='skyblue', alpha=0.4)  # Área bajo la curva
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('Función de Densidad de Probabilidad (FDP)', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

## Variable aleatoria Normal Continua
---
La distribución normal es una distribución simétrica en forma de campana. Su función de densidad de probabilidad (FDP) está dada por:
$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e{-\frac{(x-\mu)2}{2\sigma^2}}$$

**Esperanza:**
$$𝐸[𝑋]=𝜇$$

**Varianza:**
$$Var(𝑋)=𝜎^2$$

**Ejemplo**

Supongamos que $𝑋∼𝑁(70,5^2)$. Queremos calcular la probabilidad de que
𝑋 esté entre 65 y 75.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu = 70
sigma = 5

# Valores de x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
f_x = norm.pdf(x, mu, sigma)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(x, f_x, label='FDP', color='blue')
plt.fill_between(x, f_x, where=(x >= 65) & (x <= 75), color='skyblue', alpha=0.5, label='P(65 ≤ X ≤ 75)')
plt.title('Distribución Normal', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()

# Probabilidad
probabilidad = norm.cdf(75, mu, sigma) - norm.cdf(65, mu, sigma)
print(f"P(65 ≤ X ≤ 75): {probabilidad:.4f}")
```

## Variable aleatoria Exponencial Continua
---
La distribución exponencial modela el tiempo entre eventos en un proceso Poisson. Su FDP es:

$$𝑓(𝑥)=𝜆𝑒^{−𝜆𝑥},𝑥≥0$$
Donde:
𝜆: Tasa de ocurrencia.

**Esperanza:**

$$𝐸[𝑋]= \frac{1}{𝜆}$$

**Varianza:**

$$Var(𝑋)= \frac{1}{𝜆^2}$$

**Ejemplo**

Supongamos que $𝑋∼Exp(0.5)$. Queremos calcular la probabilidad de que $𝑋≤3$

```python
from scipy.stats import expon

# Parámetro
lambda_param = 0.5

# Valores de x
x = np.linspace(0, 10, 1000)
f_x = expon.pdf(x, scale=1/lambda_param)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(x, f_x, label='FDP', color='green')
plt.fill_between(x, f_x, where=(x <= 3), color='lightgreen', alpha=0.5, label='P(X ≤ 3)')
plt.title('Distribución Exponencial', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()

# Probabilidad
probabilidad = expon.cdf(3, scale=1/lambda_param)
print(f"P(X ≤ 3): {probabilidad:.4f}")
```

## Variable aleatoria Gamma Continua
---
La distribución gamma modela tiempos de espera acumulativos. Su FDP es:
$$f(x) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{\Gamma(k)}, \quad x \geq 0$$

Donde:

𝑘: Forma.

𝜆: Tasa.

**Esperanza:**
$$𝐸[𝑋]=\frac{𝑘}{𝜆}$$

**Varianza:**
$$Var(𝑋)= \frac{𝑘}{𝜆^2}$$

**Ejemplo**

Supongamos que $𝑋∼Gamma(𝑘=2,𝜆=1)$. Queremos calcular la probabilidad de que
$𝑋≤4$.

```python
from scipy.stats import gamma

# Parámetros
k = 2
lambda_param = 1

# Valores de x
x = np.linspace(0, 10, 1000)
f_x = gamma.pdf(x, k, scale=1/lambda_param)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(x, f_x, label='FDP', color='purple')
plt.fill_between(x, f_x, where=(x <= 4), color='violet', alpha=0.5, label='P(X ≤ 4)')
plt.title('Distribución Gamma', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()

# Probabilidad
probabilidad = gamma.cdf(4, k, scale=1/lambda_param)
print(f"P(X ≤ 4): {probabilidad:.4f}")
```

## Variable aleatoria Beta Continua
---
La distribución beta modela proporciones. Su FDP es:
$$f(x) = \frac{x{\alpha-1}(1-x){\beta-1}}{B(\alpha, \beta)}, \quad 0 \leq x \leq 1$$

Donde:

- 𝛼,𝛽: Parámetros de forma.
- 𝐵(𝛼,𝛽): Función beta.

**Esperanza:**
$$𝐸[𝑋]= \frac{𝛼}{𝛼+𝛽}$$

**Varianza:**
$$Var(𝑋)=\frac{𝛼𝛽}{(𝛼+𝛽)^2(𝛼+𝛽+1)}$$

**Ejemplo**

Supongamos que $𝑋∼Beta(𝛼=2,𝛽=5)$. Queremos calcular la probabilidad de que
$0.2≤𝑋≤0.6$

```python
from scipy.stats import beta

# Parámetros
alpha = 2
beta_param = 5

# Valores de x
x = np.linspace(0, 1, 1000)
f_x = beta.pdf(x, alpha, beta_param)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(x, f_x, label='FDP', color='orange')
plt.fill_between(x, f_x, where=(x >= 0.2) & (x <= 0.6), color='gold', alpha=0.5, label='P(0.2 ≤ X ≤ 0.6)')
plt.title('Distribución Beta', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()

# Probabilidad
probabilidad = beta.cdf(0.6, alpha, beta_param) - beta.cdf(0.2, alpha, beta_param)
print(f"P(0.2 ≤ X ≤ 0.6): {probabilidad:.4f}")
```

# Simulación de Variables Aleatorias
---

La simulación de variables aleatorias es una técnica fundamental en la modelación de sistemas estocásticos. Permite generar valores que siguen una distribución de probabilidad específica, lo cual es esencial para analizar y predecir el comportamiento de sistemas complejos en diversas áreas como la ingeniería, economía, ciencias sociales y más.

**Pasos para Simular Variables Aleatorias**
1. **Generación de Números Aleatorios Uniformes:**

Se generan números aleatorios $𝑈$ distribuidos uniformemente en el intervalo
$[0,1]$.

Estos números son la base para transformar a otras distribuciones.

2. **Transformación a la Distribución Deseada:**

Se utiliza una técnica de transformación para convertir los números uniformes en valores que sigan la distribución deseada.

Ejemplo: Método de la Transformada Inversa.

3. **Validación:**

Se verifica que los valores generados sigan la distribución esperada mediante pruebas estadísticas como la prueba de Kolmogorov-Smirnov.


**Métodos Comunes de Simulación**

**1. Método de la Transformada Inversa:**

- Se utiliza para distribuciones cuya función de distribución acumulativa (FDA) tiene una inversa analítica.
- Fórmula: $𝑋=𝐹−1(𝑈)$, donde $𝑈$ es un número aleatorio uniforme.

**Ejemplo**

Queremos simular una variable aleatoria exponencial con parámetro
𝜆 = 2.

La función de distribución acumulativa (FDA) de la exponencial es:
$$𝐹(𝑥)=1−𝑒^{−𝜆𝑥},𝑥≥0$$

Invertimos la FDA para obtener:
$$F^{−1}(𝑢)=−\frac{ln(1−𝑢)}{𝜆}$$

Donde 𝑢 es un número aleatorio uniforme en [0,1].

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro
lambda_param = 2

# Generación de números aleatorios
U = np.random.uniform(0, 1, 10000)
X = -np.log(1 - U) / lambda_param

# Gráfica
plt.hist(X, bins=50, density=True, alpha=0.7, color='skyblue', label='Simulación')
x = np.linspace(0, 5, 1000)
f_x = lambda_param * np.exp(-lambda_param * x)
plt.plot(x, f_x, color='red', label='FDP teórica')
plt.title('Método de la Transformada Inversa')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

**2. Método de Aceptación y Rechazo:**

- Se utiliza cuando la FDA inversa no es fácil de calcular.
- Genera valores candidatos y los acepta o rechaza según una función de densidad de probabilidad.

**Ejemplo**

Queremos simular una variable aleatoria con distribución triangular en [0,1], cuya FDP es:
$$𝑓(𝑥)=2𝑥,0≤𝑥≤1$$

Elegimos una función de referencia $𝑔(𝑥)$ que sea fácil de simular, como la distribución uniforme en $[0,1]$.

Generamos un valor candidato 𝑦 de $𝑔(𝑥)$ y un número uniforme 𝑢 en $[0,1]$.

Aceptamos 𝑦 si $𝑢≤\frac{𝑓(𝑦)}{𝑔(𝑦)}$.

```python
# Generación de variables aleatorias
n = 10000
samples = []
while len(samples) < n:
    y = np.random.uniform(0, 1)  # Candidato
    u = np.random.uniform(0, 1)  # Uniforme
    if u <= 2 * y:  # Condición de aceptación
        samples.append(y)

# Gráfica
plt.hist(samples, bins=50, density=True, alpha=0.7, color='skyblue', label='Simulación')
x = np.linspace(0, 1, 1000)
f_x = 2 * x
plt.plot(x, f_x, color='red', label='FDP teórica')
plt.title('Método de Aceptación y Rechazo')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

**3. Método de Box-Muller:**

- Específico para generar variables aleatorias normales a partir de números uniformes.

**Ejemplo**

Queremos simular dos variables aleatorias independientes con distribución normal estándar $(𝜇=0, 𝜎2=1)$.

Generamos dos números aleatorios uniformes $𝑢_1$ y $𝑢_2$ en $[0,1]$.

Calculamos:
$$
Z_0 = \sqrt{-2 \ln(u_1)} \cos(2\pi u_2)
$$
$$
Z_1 = \sqrt{-2 \ln(u_1)} \sin(2\pi u_2)
$$

```python
# Generación de números aleatorios
u1 = np.random.uniform(0, 1, 5000)
u2 = np.random.uniform(0, 1, 5000)
z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
z1 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

# Gráfica
plt.hist(z0, bins=50, density=True, alpha=0.7, color='skyblue', label='Simulación')
x = np.linspace(-4, 4, 1000)
f_x = (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)
plt.plot(x, f_x, color='red', label='FDP teórica')
plt.title('Método de Box-Muller')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

# Función Generadora de Momentos (FGM)
---
La Función Generadora de Momentos (FGM) es una herramienta matemática utilizada en probabilidad y estadística para describir las propiedades de una variable aleatoria.

**Definición:**

La FGM de una variable aleatoria 𝑋 se define como:
$$𝑀_𝑋(𝑡)=𝐸[𝑒^{𝑡𝑋}],𝑡∈𝑅$$
donde:

𝐸 es el valor esperado.

𝑡 es un parámetro real.

La FGM existe siempre que la esperanza $𝐸[𝑒^{𝑡𝑋}]$ sea finita para algún intervalo de valores de 𝑡.

**Propiedades:**

1. **Generación de Momentos:** Los momentos de 𝑋 se obtienen derivando la FGM y evaluando en 𝑡=0:
$$\mathbb{E}[X^n] = M_X^{(n)}(0) = \frac{d^n M_X(t)}{dt^n} \bigg|_{t=0}$$

2. **Unicidad:** Si la FGM existe en un intervalo alrededor de 𝑡=0, determina unívocamente la distribución de 𝑋.

3. **Relación con la Transformada de Laplace:** Para variables aleatorias continuas, la FGM es equivalente a la transformada bilateral de Laplace de la función de densidad de probabilidad.

**Ejemplo:**

Supongamos que 𝑋 es una variable aleatoria que sigue una distribución exponencial con parámetro 𝜆=2. Queremos calcular la FGM y usarla para obtener la media y la varianza.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial
lambda_param = 2

# Definimos la variable simbólica t
t = sp.symbols('t')

# Función Generadora de Momentos (FGM) para la distribución exponencial
M_X = lambda_param / (lambda_param - t)

# Derivadas para calcular momentos
M_X_prime = sp.diff(M_X, t)  # Primera derivada
M_X_double_prime = sp.diff(M_X_prime, t)  # Segunda derivada

# Evaluamos la media (primer momento)
media = M_X_prime.subs(t, 0)
print(f"Media (E[X]): {media}")

# Evaluamos la varianza (segundo momento menos el cuadrado de la media)
varianza = M_X_double_prime.subs(t, 0) - media**2
print(f"Varianza (Var[X]): {varianza}")

# Gráfica de la FGM
t_vals = np.linspace(-1, lambda_param - 0.1, 500)  # Valores de t (evitamos t >= lambda)
M_X_vals = [lambda_param / (lambda_param - t_val) for t_val in t_vals]

plt.figure(figsize=(8, 5))
plt.plot(t_vals, M_X_vals, label='FGM: $M_X(t)$', color='blue')
plt.axvline(x=lambda_param, color='red', linestyle='--', label=r'$t = \lambda$ (límite)')
plt.title('Función Generadora de Momentos (FGM) para la Exponencial', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('$M_X(t)$', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

# Función Característica (FC)
---
La Función Característica (FC) es una herramienta matemática utilizada en probabilidad y estadística para describir la distribución de una variable aleatoria. Es especialmente útil porque siempre existe, incluso en casos donde la función generadora de momentos no está definida. Además, permite calcular momentos, analizar distribuciones y demostrar teoremas fundamentales como el Teorema Central del Límite.

**Definición:**

La función característica de una variable aleatoria 𝑋 se define como:
$$𝜑_𝑋(𝑡)=𝐸[𝑒^{𝑖𝑡𝑋}]$$

donde:

𝐸 es el valor esperado.

𝑖 es la unidad imaginaria $(𝑖^2=−1)$.

𝑡 es un número real.

Para variables aleatorias continuas con función de densidad 𝑓(𝑥), la FC se expresa como:
$$\varphi_X(t) = \int_{-\infty}^{\infty} e^{itx} f(x) \, dx$$

Para variables aleatorias discretas con función de probabilidad
$𝑃(𝑋=𝑥_𝑖)=𝑝_𝑖$, la FC es:
$$\varphi_X(t) = \sum_{i} e^{itx_i} p_i$$

**Propiedades:**

1. **Unicidad:** La FC determina unívocamente la distribución de 𝑋.

2. **Relación con Momentos:** Los momentos de 𝑋 se obtienen derivando la FC respecto a 𝑡 y evaluando en 𝑡=0:
$$\mathbb{E}[X^n] = \frac{d^n \varphi_X(t)}{dt^n} \bigg|_{t=0}$$

3. **Transformada de Fourier:** La FC es la transformada de Fourier de la función de densidad 𝑓(𝑥).

4. **Multiplicación de Variables Independientes:** Si 𝑋 y 𝑌 son independientes, entonces:
$$\varphi_{X+Y}(t) = \varphi_X(t) \cdot \varphi_Y(t)$$

**Ejemplo:**

La función característica de una variable aleatoria 𝑋∼𝑁(0,1) es:
$$\varphi_X(t) = e{-\frac{t2}{2}}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de t
t_vals = np.linspace(-10, 10, 1000)

# Función característica para la distribución normal estándar
phi_X = np.exp(-t_vals**2 / 2)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(t_vals, phi_X, label=r'$\varphi_X(t) = e^{-t^2 / 2}$', color='blue')
plt.title('Función Característica (FC) de la Distribución Normal Estándar', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel(r'$\varphi_X(t)$', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

# Ley de los grandes números (LGN)
---
La Ley de los Grandes Números (LGN) es un teorema fundamental en la teoría de la probabilidad que describe el comportamiento del promedio de una sucesión de variables aleatorias conforme aumenta el número de ensayos. Es esencial para garantizar resultados estables a largo plazo en experimentos aleatorios.

**Definición:**

La LGN establece que, bajo ciertas condiciones, el promedio de una muestra de variables aleatorias independientes e idénticamente distribuidas (iid) converge al valor esperado de la población cuando el tamaño de la muestra tiende a infinito.

Existen dos versiones principales de la LGN:

1. **LGN Débil:** El promedio de las variables aleatorias converge en probabilidad al valor esperado. Es decir:
$$\lim_{n \to \infty} P\left(\left| \frac{1}{n} \sum_{i=1}^n X_i - \mathbb{E}[X] \right| > \epsilon \right) = 0, \quad \forall \epsilon > 0$$

2. **LGN Fuerte:** El promedio de las variables aleatorias converge casi seguramente al valor esperado. Es decir:
$$P\left(\lim_{n \to \infty} \frac{1}{n} \sum_{i=1}^n X_i = \mathbb{E}[X] \right) = 1$$

**Ejemplo:**

Supongamos que lanzamos una moneda justa repetidamente. La probabilidad de obtener cara en un lanzamiento es 𝑝=0.5. Según la LGN:

Si lanzamos la moneda un número pequeño de veces, el promedio de caras puede desviarse significativamente de 0.5.

Si lanzamos la moneda un número muy grande de veces, el promedio de caras se acercará cada vez más a 0.5.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de lanzamientos de una moneda justa
n_lanzamientos = 10000
resultados = np.random.choice([0, 1], size=n_lanzamientos)  # 0: cruz, 1: cara
promedios = np.cumsum(resultados) / np.arange(1, n_lanzamientos + 1)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(promedios, label='Promedio acumulado', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='Valor esperado (0.5)')
plt.title('Ley de los Grandes Números (LGN)', fontsize=14)
plt.xlabel('Número de lanzamientos', fontsize=12)
plt.ylabel('Promedio de caras', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

# Teorema central del límite (TCL)
---
El Teorema Central del Límite (TCL) es uno de los resultados más importantes en la teoría de la probabilidad y la estadística. Establece que, bajo ciertas condiciones, la distribución de la suma o promedio de un conjunto de variables aleatorias independientes e idénticamente distribuidas (iid) tiende a una distribución normal conforme el tamaño de la muestra aumenta.

**Definición:**

Sea $𝑋_1,𝑋_2,…,𝑋_𝑛$ un conjunto de variables aleatorias iid con:

Media $𝜇=𝐸[𝑋_𝑖]$

Varianza $𝜎^2=Var(𝑋_𝑖)$

El TCL establece que la distribución de la variable aleatoria:
$$Z_n = \frac{\sum_{i=1}^n X_i - n\mu}{n\sigma}$$

converge en distribución a una normal estándar 𝑁(0,1) cuando $𝑛→∞$.

**Condiciones:**

1. Las variables aleatorias deben ser independientes.
2. Las variables deben tener la misma distribución (iid).
3. La media y la varianza deben ser finitas.

**Ejemplo:**

Supongamos que queremos analizar la altura promedio de estudiantes en una universidad. Aunque la distribución de alturas individuales no sea perfectamente normal, el TCL asegura que la distribución de la media muestral será aproximadamente normal si el tamaño de la muestra es suficientemente grande.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de variables aleatorias
n_muestras = 1000  # Número de muestras
n_variables = 50   # Tamaño de cada muestra

# Generamos variables aleatorias de una distribución uniforme
datos = np.random.uniform(0, 10, (n_muestras, n_variables))

# Calculamos las medias muestrales
medias_muestrales = np.mean(datos, axis=1)

# Gráfica del histograma
plt.figure(figsize=(8, 5))
plt.hist(medias_muestrales, bins=30, density=True, alpha=0.7, color='skyblue', label='Medias muestrales')
plt.title('Teorema Central del Límite (TCL)', fontsize=14)
plt.xlabel('Media muestral', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
