## Variable aleatoria Uniforme Discreta

Una variable aleatoria $X$ tiene distribuci贸n uniforme discreta en el conjunto si su funci贸n de masa de probabilidades est谩 dada por
$$f_{X}(k)=\mathbb{P}(X=k)=\begin{cases}

 $$X\sim Unif(x_{1},\dots,x_{n}).$$

Se llama uniforme porque cada uno de los posibles valores de $X$ tienen la misma probabilidad.

	$$\mathbb{E}[X] = \sum_{k}kf_{X}(k)=\frac{x_n + x_1}{2}.$$

Cada participante tiene el mismo chance de ser seleccionado en cualquier posici贸n (del 1 al 8). Si definimos la variable aleatoria  como la posici贸n asignada a un jugador espec铆fico, entonces:

## Variable aleatoria Bernoulli Discreta
---
El modelo probabil铆stico Bernoulli se aplica a un experimento cuyo espacio muestral est谩 constituido s贸lo por dos resultados posibles, 茅xito y fracaso:

Se considerar谩 una v.a $X$ sobre el espacio muestral $$\Omega=\{\text{茅xito, fracaso}\}$$
de tal forma que
* $X(\{\text{茅xito}\}) = 1$
* $X(\{\text{fracaso}\}) = 0$

Las probabilidades asociadas a este modelo son:

* $\mathbb{P}(\{\text{茅xito}\}) = p$
* $\mathbb{P}(\{\text{fracaso}\}) = 1-p$

donde $0<p<1$.


Entonces la funci贸n de densidad de masa de probabilidades de $X$ est谩 dada por:

$$f_{X}(k)=\begin{cases}
p &\quad \text{si } k=1 \\
1-p &\quad \text{si } k=0
\end{cases}$$

Tenemos que la notaci贸n es la siguiente:

 $$X\sim \text{Ber}(p)$$

 Si $X\sim \text{Ber}(p)$, entonces su esperanza es:
\begin{align*}
\mathbb{E}[X] = \sum_{k=0}^{1}kf_{X}(k) = 1\cdot p + 0\cdot (1-p) = p.
\end{align*}

Y su varianza se calcula usando,
$$\mathbb{E}[X^{2}] = \sum_{k=0}^{1}k^{2}f_{X}(k) = 1^{2}\cdot p + 0^{2}\cdot (1-p) = p.$$
Por lo tanto,
$$\text{Var}(X)=\mathbb{E}[X^{2}]-(\mathbb{E}[X])^{2} = p-p^{2} = p(1-p)$$

**Esperanza matem谩tica**

$$Var() = (1) = 0.4  0.6 = 0.24$$

## Variable aleatoria Poisson Discreta
---
Esta variable aleatoria toma los valores sobre el conjunto $\{0,1,2,\dots\}$ y tiene un par谩metro $\lambda>0$, el cual representa el n煤mero de veces que se espera que ocurra un evento durante un intervalo dado.

La funci贸n de masa de probabilidades de una variable aleatoria Poisson, $X$, est谩 dada por:
$$f_{X}(k)=\begin{cases}
	e^{-\lambda}\frac{\lambda^{k}}{k!} &\quad \text{si } k=0,1,2,\dots \\
	0 &\quad \text{en otro caso}  
\end{cases}$$

En efecto, es una funci贸n de masa de probabilidades debido a que
$$\begin{align*}
\sum_{k=0}^{\infty}e^{-\lambda}\frac{\lambda^{k}}{k!} &= e^{-\lambda}\sum_{k=0}^{\infty}\frac{\lambda^{k}}{k!} \\ &= e^{-\lambda}\underbrace{\left(1+\lambda+\frac{\lambda^{2}}{2!}+\frac{\lambda^{3}}{3!}+\cdots\right)}_{\text{desarrollo de Taylor de }e^{\lambda}} \\ &= e^{-\lambda}e^{\lambda}=1.
\end{align*}$$

**Ejemplo**

Un centro de llamadas recibe, en promedio, 5 llamadas por minuto. Queremos calcular la probabilidad de que el centro reciba exactamente 3 llamadas en un minuto.

*F贸rmula de la distribuci贸n de Poisson:*

La probabilidad de que ocurran  eventos en un intervalo dado est谩 dada por:

$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$
donde:

 es el n煤mero promedio de eventos (en este caso, 5 llamadas por minuto),

 es el n煤mero de eventos que queremos calcular (en este caso, 3 llamadas),

 es la base del logaritmo natural (2.718).

Sustituyendo los valores:
$$P(X = 3) = \frac{5^3 e^{-5}}{3!} = \frac{125 \cdot e^{-5}}{6}
$$


$$P(X = 3) \approx \frac{125 \cdot 0.0067}{6} \approx 0.1404
$$

**Esperanza matem谩tica y varianza:**

Ambas son iguales a . En este caso,
$$[]=Var()=5$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Par谩metro de la distribuci贸n Poisson
lambda_poisson = 5  # Promedio de llamadas por minuto

# Valores posibles de la variable aleatoria
x = np.arange(0, 15)  # Rango de valores para X (0 a 14)

# Funci贸n de masa de probabilidad (PMF)
pmf = poisson.pmf(x, lambda_poisson)

# Mostrar resultados para un valor espec铆fico
k = 3  # N煤mero de llamadas
prob_k = poisson.pmf(k, lambda_poisson)
print(f"P(X = {k}): {prob_k:.4f}")

# Gr谩fica de la PMF
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Funci贸n de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('N煤mero de llamadas (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Variable aleatoria Geom茅trica Discreta
---
El modelo probabil铆stico geom茅trico se aplica a un experimento que consiste en una secuencia de ensayos independientes, donde cada ensayo tiene dos posibles resultados: 茅xito y fracaso.

Se considerar谩 una variable aleatoria  sobre el espacio muestral:
$$\Omega = \{ \text{primer 茅xito en el ensayo } k \mid k \in \mathbb{N} \}
$$
De tal forma que:
$$X = k \quad \text{si el primer 茅xito ocurre en el ensayo } k.
$$
Las probabilidades asociadas a este modelo son:
$$P(X = k) = p(1-p)^{k-1}, \quad \text{para } k = 1, 2, 3, \dots
$$
donde
$0<<1$

Entonces, la funci贸n de masa de probabilidad de  est谩 dada por:
$$f_X(k) =
\begin{cases}
p(1-p)^{k-1} & \text{si } k \geq 1, \\

\end{cases}
$$

Tenemos que la notaci贸n es la siguiente:
$$Geo()$$

**Esperanza matem谩tica:**
$$\mathbb{E}[X] = \sum_{k=1}^{\infty} k \cdot f_X(k) = \sum_{k=1}^{\infty} k \cdot p(1-p)^{k-1}.
$$

Usando propiedades de series geom茅tricas, se obtiene:
$$\mathbb{E}[X] = \frac{1}{p}.$$

**Varianza:**
$$\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2.$$

$$\mathbb{E}[X^2] = \sum_{k=1}^{\infty} k^2 \cdot f_X(k).$$

Usando propiedades de series geom茅tricas, se obtiene:
$$\text{Var}(X) = \frac{1-p}{p^2}.$$

**Ejemplo**

**Funci贸n de masa de probabilidad:**

$$P(X = k) = p(1-p)^{k-1}, \quad \text{para } k = 1, 2, 3, \dots$$

**Probabilidad de pasar en el primer intento ( = 1):**
$$P(X = 1) = 0.3(1-0.3)^{1-1} = 0.3$$

**Probabilidad de pasar en el segundo intento ( = 2):**
$$P(X = 2) = 0.3(1-0.3)^{2-1} = 0.3 \cdot 0.7 = 0.21$$

**Probabilidad de pasar en el tercer intento ( = 3):**
$$P(X = 3) = 0.3(1-0.3)^{3-1} = 0.3 \cdot 0.7^2 = 0.3 \cdot 0.49 = 0.147$$

$$\mathbb{E}[X] = \frac{1}{p} = \frac{1}{0.3} \approx 3.33$$

**Varianza:**
$$\text{Var}(X) = \frac{1-p}{p^2} = \frac{1-0.3}{0.3^2} = \frac{0.7}{0.09} \approx 7.78$$

## Variable aleatoria Binomial Negativa Discreta
---
El modelo probabil铆stico de la binomial negativa se aplica a un experimento que consiste en una secuencia de ensayos independientes, donde cada ensayo tiene dos posibles resultados: 茅xito y fracaso. Este modelo se utiliza para determinar el n煤mero de fracasos antes de alcanzar un n煤mero fijo de 茅xitos.

Se considerar谩 una variable aleatoria $$ sobre el espacio muestral:
$$\Omega = \{\text{n煤mero de fracasos antes de alcanzar } r \text{ 茅xitos}\}$$

De tal forma que:
$$X = k \quad \text{si se producen } k \text{ fracasos antes de alcanzar } r \text{ 茅xitos}.$$

Las probabilidades asociadas a este modelo son:
$$P(X = k) = \binom{k+r-1}{r-1} p^r (1-p)^k, \quad \text{para } k = 0, 1, 2, \dots$$


donde:

 es el n煤mero fijo de 茅xitos,

 es la probabilidad de 茅xito en cada ensayo,

 es el n煤mero de fracasos antes de alcanzar  茅xitos,

$\binom{k+r-1}{r-1}$ es el coeficiente binomial.

Tenemos que la notaci贸n es la siguiente:
$$NB(,)$$

**Esperanza matem谩tica:**
$$\mathbb{E}[X] = \frac{r(1-p)}{p}.$$

**Varianza:**
$$\text{Var}(X) = \frac{r(1-p)}{p^2}.$$

**Ejemplo**

La probabilidad de que el vendedor realice  = 4 llamadas fallidas est谩 dada por:

$$P(X = 4) = \binom{4+3-1}{3-1} (0.25)^3 (1-0.25)^4$$

Sustituyendo los valores:

$$P(X = 4) = \binom{6}{2} (0.25)^3 (0.75)^4$$

Calculamos el coeficiente binomial:

$$\binom{6}{2} = \frac{6!}{2!(6-2)!} = \frac{6 \cdot 5}{2 \cdot 1} = 15$$

Calculamos las potencias:

$$(0.25)^3 = 0.015625, \quad (0.75)^4 = 0.31640625$$

Sustituimos todo:

$$P(X = 4) = 15 \cdot 0.015625 \cdot 0.31640625 = 0.0741$$

**Esperanza matem谩tica:**

```python
import numpy as np
from scipy.stats import nbinom
import matplotlib.pyplot as plt

# Par谩metros de la distribuci贸n binomial negativa
r = 3  # N煤mero de 茅xitos
p = 0.25  # Probabilidad de 茅xito

# Valores posibles de la variable aleatoria
x = np.arange(0, 20)  # N煤mero de fracasos (k)

# Funci贸n de masa de probabilidad (PMF)
pmf = nbinom.pmf(x, r, p)

# Funci贸n de distribuci贸n acumulativa (CDF)
cdf = nbinom.cdf(x, r, p)

# Mostrar resultados para un valor espec铆fico
k = 4  # N煤mero de fracasos
prob_k = nbinom.pmf(k, r, p)
print(f"P(X = {k}): {prob_k:.4f}")

# Gr谩fica de la PMF
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='PMF')
plt.title('Funci贸n de Masa de Probabilidad (PMF)', fontsize=14)
plt.xlabel('N煤mero de fracasos (X)', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Gr谩fica de la CDF
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post', color='orange', alpha=0.8, label='CDF')
plt.title('Funci贸n de Distribuci贸n Acumulativa (CDF)', fontsize=14)
plt.xlabel('N煤mero de fracasos (X)', fontsize=12)
plt.ylabel('Probabilidad acumulada', fontsize=12)
plt.xticks(x)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Existen dos versiones principales de la LGN:

1. **LGN D茅bil:** El promedio de las variables aleatorias converge en probabilidad al valor esperado. Es decir:
$$\lim_{n \to \infty} P\left(\left| \frac{1}{n} \sum_{i=1}^n X_i - \mathbb{E}[X] \right| > \epsilon \right) = 0, \quad \forall \epsilon > 0$$

2. **LGN Fuerte:** El promedio de las variables aleatorias converge casi seguramente al valor esperado. Es decir:
$$P\left(\lim_{n \to \infty} \frac{1}{n} \sum_{i=1}^n X_i = \mathbb{E}[X] \right) = 1$$

**Ejemplo:**

Supongamos que lanzamos una moneda justa repetidamente. La probabilidad de obtener cara en un lanzamiento es =0.5. Seg煤n la LGN:

Si lanzamos la moneda un n煤mero peque帽o de veces, el promedio de caras puede desviarse significativamente de 0.5.

Si lanzamos la moneda un n煤mero muy grande de veces, el promedio de caras se acercar谩 cada vez m谩s a 0.5.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulaci贸n de lanzamientos de una moneda justa
n_lanzamientos = 10000
resultados = np.random.choice([0, 1], size=n_lanzamientos)  # 0: cruz, 1: cara
promedios = np.cumsum(resultados) / np.arange(1, n_lanzamientos + 1)

# Gr谩fica
plt.figure(figsize=(8, 5))
plt.plot(promedios, label='Promedio acumulado', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='Valor esperado (0.5)')
plt.title('Ley de los Grandes N煤meros (LGN)', fontsize=14)
plt.xlabel('N煤mero de lanzamientos', fontsize=12)
plt.ylabel('Promedio de caras', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

# Teorema central del l铆mite (TCL)
---
El Teorema Central del L铆mite (TCL) es uno de los resultados m谩s importantes en la teor铆a de la probabilidad y la estad铆stica. Establece que, bajo ciertas condiciones, la distribuci贸n de la suma o promedio de un conjunto de variables aleatorias independientes e id茅nticamente distribuidas (iid) tiende a una distribuci贸n normal conforme el tama帽o de la muestra aumenta.

**Definici贸n:**

Sea $_1,_2,,_$ un conjunto de variables aleatorias iid con:

Media $=[_]$

Varianza $^2=Var(_)$

El TCL establece que la distribuci贸n de la variable aleatoria:
$$Z_n = \frac{\sum_{i=1}^n X_i - n\mu}{n\sigma}$$

converge en distribuci贸n a una normal est谩ndar (0,1) cuando $$.

**Condiciones:**

2. Las variables deben tener la misma distribuci贸n (iid).

**Ejemplo:**

Supongamos que queremos analizar la altura promedio de estudiantes en una universidad. Aunque la distribuci贸n de alturas individuales no sea perfectamente normal, el TCL asegura que la distribuci贸n de la media muestral ser谩 aproximadamente normal si el tama帽o de la muestra es suficientemente grande.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulaci贸n de variables aleatorias
n_muestras = 1000  # N煤mero de muestras
n_variables = 50   # Tama帽o de cada muestra

# Generamos variables aleatorias de una distribuci贸n uniforme
datos = np.random.uniform(0, 10, (n_muestras, n_variables))

# Calculamos las medias muestrales
medias_muestrales = np.mean(datos, axis=1)

# Gr谩fica del histograma
plt.figure(figsize=(8, 5))
plt.hist(medias_muestrales, bins=30, density=True, alpha=0.7, color='skyblue', label='Medias muestrales')
plt.title('Teorema Central del L铆mite (TCL)', fontsize=14)
plt.xlabel('Media muestral', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
