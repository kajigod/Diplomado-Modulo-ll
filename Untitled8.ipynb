Este conjunto de datos ofrece una vista de los precios de las viviendas en la Ciudad de México (CDMX), capturada durante un período en el que el peso mexicano se cotizaba a aproximadamente 18.80 por dólar estadounidense. Proporciona una visión clara del mercado inmobiliario residencial dentro de la CDMX, un centro urbano dinámico e influyente. Los datos incluyen variables clave como el tipo de propiedad, la localidad específica dentro de la ciudad, detalles de precios y la superficie. Esta combinación permite un análisis enfocado de cómo estas características principales influyen en el valor de las propiedades en una metrópolis latinoamericana importante bajo condiciones económicas específicas.

* Usar PCA para reducir la redundancia entre variables de precios y superficies.

* Aplicar Clustering sobre los componentes principales (o directamente sobre variables normalizadas) para segmentar las propiedades.

Se obtendrá:

* Una visión compacta y explicativa de las relaciones entre variables (PCA).

* Una segmentación útil para interpretar los distintos tipos de inmuebles (Clusters).

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#  Selección de variables numéricas
num_vars = [
    "price",
    "price_aprox_local_currency",
    "price_aprox_usd",
    "surface_total_in_m2",
    "surface_covered_in_m2",
    "price_usd_per_m2",
    "price_per_m2"
]

X = df[num_vars].dropna()  # eliminar posibles nulos

# Normalización
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Varianza explicada
explained_var = pca.explained_variance_ratio_

print("Varianza explicada por cada componente:")
for i, var in enumerate(explained_var, 1):
    print(f"CP{i}: {var:.4f} ({var*100:.2f}%)")

print("\nVarianza acumulada:", explained_var.cumsum())

# Gráfica de codo (Scree Plot)
plt.figure(figsize=(6,4))
plt.plot(range(1, len(explained_var)+1), explained_var.cumsum(), marker='o')
plt.xlabel("Número de componentes")
plt.ylabel("Varianza explicada acumulada")
plt.title("Gráfica de codo - PCA")
plt.grid(True)
plt.show()

# Loadings (contribución de variables)
loadings = pd.DataFrame(
    pca.components_.T,
    index=num_vars,
    columns=[f"CP{i+1}" for i in range(len(num_vars))]
)

print("\nMatriz de loadings (primeros 3 componentes):")
print(loadings.iloc[:, :3])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Usar los datos del PCA
X_cluster = X_pca[:, :3]   # primeras 3 componentes principales

# Método del codo y silhouette
inertia = []
silhouette = []
K = range(2, 8)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_cluster)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_cluster, kmeans.labels_))

# Gráfica del codo
plt.figure(figsize=(6,4))
plt.plot(K, inertia, marker="o")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Inercia (Suma de distancias)")
plt.title("Método del codo - KMeans (3 CPs)")
plt.show()

# Gráfica de Silhouette
plt.figure(figsize=(6,4))
plt.plot(K, silhouette, marker="o", color="orange")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Evaluación con Silhouette (3 CPs)")
plt.show()

# Entrenar modelo final con k=3 (ejemplo)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["cluster"] = kmeans.fit_predict(X_cluster)

print("\nCantidad de propiedades por cluster:")
print(df["cluster"].value_counts())

# Visualización 3D de clusters
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")

scatter = ax.scatter(
    X_cluster[:,0], X_cluster[:,1], X_cluster[:,2],
    c=df["cluster"], cmap="Set2", s=50
)

ax.set_xlabel("CP1")
ax.set_ylabel("CP2")
ax.set_zlabel("CP3")
ax.set_title("Clustering con 3 CPs")

plt.legend(*scatter.legend_elements(), title="Cluster")
plt.show()
```
