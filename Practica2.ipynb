5. Determina si eliminar los registros inconsistentes es representativo de la
poblaci√≥n. Justificar

```python
# Total de registros originales
total = len(df)

# Registros que quedar√≠an si eliminamos los que tienen valores faltantes
eliminables = len(df.dropna())

# Porcentaje de registros que se eliminar√≠an
porcentaje = 100 * (total - eliminables) / total

print(f"üîç Registros originales: {total}")
print(f"‚úÖ Registros sin valores nulos: {eliminables}")
print(f"‚ö†Ô∏è Porcentaje que se eliminar√≠a: {porcentaje:.2f}%")
```

Al evaluar la cantidad de registros con valores faltantes:

- Registros originales: 19,723  
- Registros sin valores faltantes: 19,709
- Porcentaje de eliminaci√≥n: **0.07%**

> Dado que el porcentaje de eliminaci√≥n es extremadamente bajo (<1%), **eliminar estos registros no compromete la representatividad de la muestra**. Por lo tanto, es estad√≠sticamente aceptable descartarlos sin afectar la validez del an√°lisis.

 6. Ejemplificar un proceso de transformaci√≥n para valores faltantes o datos
inconsistentes (‚Äúraros‚Äù)

```python
# Transformar valores faltantes en columnas B y C usando la mediana
df["B"] = df["B"].fillna(df["B"].median())
df["C"] = df["C"].fillna(df["C"].median())
# Verificar que no queden valores nulos en B y C
print("Valores nulos en B:", df["B"].isnull().sum())
print("Valores nulos en C:", df["C"].isnull().sum())
```

Para tratar los valores faltantes de las columnas `B` y `C`, se utiliz√≥ la **mediana** como t√©cnica de imputaci√≥n:

- La mediana es una medida robusta frente a valores at√≠picos.
- Es adecuada para variables num√©ricas con distribuci√≥n asim√©trica, como `B`.
- Esta t√©cnica conserva la cantidad total de registros y evita sesgos por eliminaci√≥n.

 7. Obtener la matriz de correlaci√≥n de hasta 8 variables. Interpretar el resultado
