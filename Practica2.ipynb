5. Determina si eliminar los registros inconsistentes es representativo de la
población. Justificar

```python
# Total de registros originales
total = len(df)

# Registros que quedarían si eliminamos los que tienen valores faltantes
eliminables = len(df.dropna())

# Porcentaje de registros que se eliminarían
porcentaje = 100 * (total - eliminables) / total

print(f"🔍 Registros originales: {total}")
print(f"✅ Registros sin valores nulos: {eliminables}")
print(f"⚠️ Porcentaje que se eliminaría: {porcentaje:.2f}%")
```

Al evaluar la cantidad de registros con valores faltantes:

- Registros originales: 19,723  
- Registros sin valores faltantes: 19,709
- Porcentaje de eliminación: **0.07%**

> Dado que el porcentaje de eliminación es extremadamente bajo (<1%), **eliminar estos registros no compromete la representatividad de la muestra**. Por lo tanto, es estadísticamente aceptable descartarlos sin afectar la validez del análisis.

 6. Ejemplificar un proceso de transformación para valores faltantes o datos
inconsistentes (“raros”)

```python
# Transformar valores faltantes en columnas B y C usando la mediana
df["B"] = df["B"].fillna(df["B"].median())
df["C"] = df["C"].fillna(df["C"].median())
# Verificar que no queden valores nulos en B y C
print("Valores nulos en B:", df["B"].isnull().sum())
print("Valores nulos en C:", df["C"].isnull().sum())
```

Para tratar los valores faltantes de las columnas `B` y `C`, se utilizó la **mediana** como técnica de imputación:

- La mediana es una medida robusta frente a valores atípicos.
- Es adecuada para variables numéricas con distribución asimétrica, como `B`.
- Esta técnica conserva la cantidad total de registros y evita sesgos por eliminación.

 7. Obtener la matriz de correlación de hasta 8 variables. Interpretar el resultado

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar 8 variables numéricas del DataFrame
variables_8 = df.select_dtypes(include='number').iloc[:, :8]

# Calcular matriz de correlación
matriz_corr = variables_8.corr()

# Visualización con mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("🔗 Matriz de correlación (8 primeras variables numéricas)")
plt.show()
```

Se calculó la matriz de correlación entre las variables `E1` a `E8`, revelando relaciones significativas entre varios pares. A continuación se interpretan los resultados más relevantes:

#### 🔹 Correlaciones altas y positivas:
- **E5 y E7**: 0.63  
- **E3 y E5**: 0.59  
- **E1 y E7**: 0.59  
- **E2 y E6**: 0.57  
- **E3 y E7**: 0.58  

Estas correlaciones indican que estas variables tienen una **relación directa fuerte**. Es probable que estén evaluando aspectos similares, lo que sugiere una posible redundancia si se usaran todas juntas sin reducción de dimensionalidad. Podrían formar parte de un mismo componente si se aplicara un análisis factorial.

- **E2 y E5**: –0.54  
- **E2 y E1**: –0.42  
- **E4 y E1**: –0.48

Estas correlaciones negativas reflejan una **relación inversa**: cuando una variable aumenta, la otra tiende a disminuir. Esto podría deberse a que están evaluando dimensiones opuestas de un mismo constructo (por ejemplo, confianza vs. desconfianza).

#### 🟡 Otras observaciones:
- La mayoría de las variables presentan correlaciones moderadas o fuertes (positivas o negativas), lo que sugiere que estas 8 variables no son completamente independientes entre sí.
- **E8** muestra correlaciones más bajas, lo cual indica que podría estar evaluando un aspecto distinto respecto a las demás variables del grupo.

> En conjunto, el análisis sugiere que varias variables están relacionadas y podrían reducirse a menos dimensiones latentes. Este resultado es útil para aplicar técnicas como análisis factorial, clustering o selección de variables en modelos predictivos.
