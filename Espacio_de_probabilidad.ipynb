# Introducción
---
El espacio de probabilidad es una herramienta matemática que nos permite analizar situaciones donde interviene el azar. Imagina que realizamos un experimento, como lanzar una moneda o un dado. El espacio de probabilidad nos ayuda a organizar y describir todos los resultados posibles, los eventos que pueden ocurrir y las probabilidades asociadas a cada uno de ellos. Este concepto es esencial para entender y resolver problemas de incertidumbre en áreas como la estadística, la ciencia y la vida cotidiana.

Existen varios enfoques para entender y trabajar con espacios de probabilidad. Estos los vamos a estudiar más adelante, exponiendo los siguientes:
* Clásico
* Geométrico
* Frecuentista

# Enfoque clásico
---

Este es el más intuitivo y se usa cuando todos los resultados posibles son igualmente probables. Por ejemplo, al lanzar un dado, cada cara tiene la misma probabilidad de aparecer. Aquí se calcula la probabilidad como:

$$
P(A) = \frac{\text{Número de casos favorables}}{\text{Número total de casos posibles}}
$$

**Teorema (Enfoque Clásico de la Probabilidad):**

Para cualquier evento 𝐴 en un espacio muestral finito 𝑆 con resultados equiprobables, la probabilidad de 𝐴 está dada por:

$$
P(A) = \frac{|A|}{|S|}
$$

donde:

$|A|$ es el número de casos favorables al evento 𝐴

$|S|$ es el número total de resultados en el espacio muestral 𝑆

El enfoque clásico de la probabilidad tiene fundamentos sólidos en la teoría matemática y lógica. Se basa en los siguientes principios y conceptos:

**Principio de equiprobabilidad**

El enfoque clásico asume que todos los resultados del espacio muestral son igualmente probables. Este principio es válido en situaciones donde no hay razones para pensar que un resultado sea más probable que otro (por ejemplo, en el lanzamiento de un dado justo).

**Definición axiomática**

El enfoque clásico es coherente con los axiomas de Kolmogórov, que son la base matemática de la teoría de probabilidades. Estos axiomas garantizan que la probabilidad sea un valor entre 0 y 1, y que la suma de las probabilidades de todos los resultados en el espacio muestral sea igual a 1.

**Aplicaciones prácticas**

Este enfoque se aplica principalmente en problemas de juegos de azar y situaciones idealizadas donde las condiciones son controladas y los resultados son discretos y finitos.

**Relación con el concepto de proporción**

La fórmula clásica $P(A) = \frac{casos favorables}{casos posibles}$ utiliza el concepto de proporción, que es fundamental en matemáticas y lógica.

Para este enfoque vamos a revisar algunos conceptos clave, como por ejemplo:

**Espacio muestral (Ω)**

Es el conjunto que incluye todos los posibles resultados de un experimento aleatorio. Por ejemplo, si lanzamos un dado, el espacio muestral sería:

$$
Ω = \{1, 2, 3, 4, 5, 6\}
$$

**Eventos**

Un evento es un subconjunto del espacio muestral. Representa una situación específica o un conjunto de resultados que satisfacen una condición dada. Por ejemplo, el evento "número par" al lanzar un dado sería:

$$
𝐸 = {2,4,6}
$$

Aquí tenemos ejemplos de estos dos conceptos:

```python
# Definir el espacio muestral (lanzamiento de un dado)
espacio_muestral = [1, 2, 3, 4, 5, 6]

# Definir un evento: números pares
evento_numeros_pares = [x for x in espacio_muestral if x % 2 == 0]

print("Espacio Muestral:", espacio_muestral)
print("Evento - Números pares:", evento_numeros_pares)
```

**Probabilidad condicional**

Es la probabilidad de que ocurra un evento 𝐴 dado que otro evento 𝐵 ya ha ocurrido. Se calcula mediante la fórmula:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)},
$$

donde $P(A \mid B)$ es la probabilidad de que ocurran ambos eventos simultáneamente.

```python
# Supongamos que tenemos un espacio muestral con colores y formas
espacio_muestral = [('rojo', 'círculo'), ('azul', 'círculo'), ('rojo', 'cuadrado'), ('azul', 'cuadrado')]

# Calcular la probabilidad condicional de que sea rojo dado que es un círculo
evento_circulo = [x for x in espacio_muestral if x[1] == 'círculo']
evento_rojo_y_circulo = [x for x in evento_circulo if x[0] == 'rojo']

probabilidad_condicional = len(evento_rojo_y_circulo) / len(evento_circulo)

print("Probabilidad de que sea rojo dado que es círculo:", probabilidad_condicional)
```

**Independencia**

Dos eventos 𝐴 y 𝐵 son independientes si la ocurrencia de uno no afecta la probabilidad del otro. Matemáticamente, esto se expresa como:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

```python
# Definir eventos independientes: lanzar dos dados
import random

def lanzar_dado():
    return random.randint(1, 6)

# Simulación de lanzamientos
lanzamientos = [(lanzar_dado(), lanzar_dado()) for _ in range(1000)]

# Calcular si los eventos son independientes
evento_suma_7 = [par for par in lanzamientos if sum(par) == 7]
probabilidad_suma_7 = len(evento_suma_7) / len(lanzamientos)

evento_1_en_primer_dado = [par for par in lanzamientos if par[0] == 1]
probabilidad_1_primer_dado = len(evento_1_en_primer_dado) / len(lanzamientos)

print("Probabilidad de obtener suma 7:", probabilidad_suma_7)
print("Probabilidad de obtener un '1' en el primer dado:", probabilidad_1_primer_dado)-
```

# Enfoque geométrico
---

Este enfoque es útil cuando los resultados posibles de un experimento están representados en un espacio continuo. La probabilidad se calcula como una relación entre las medidas geométricas.

La probabilidad de un evento 𝐸 en el enfoque geométrico se define como:

$$
P(E) = \frac{\text{Medida del evento (área, longitud, volumen, etc.)}}{\text{Medida del espacio total}}
$$

**Espacio total:** Representa el espacio continuo donde puede ocurrir el experimento (por ejemplo, un segmento, un plano o un volumen).

**Evento:** Es un subconjunto del espacio total que cumple ciertas condiciones (por ejemplo, un intervalo dentro de un segmento, un círculo dentro de un cuadrado, etc.).

Este enfoque es ideal para problemas donde los resultados no son discretos, sino que están distribuidos de manera continua.

Este enfoque se basa en principios matemáticos sólidos y en conceptos fundamentales de la teoría de medidas. Estos principios incluyen:

**Proporcionalidad de la medida**

La probabilidad de un evento es proporcional a la medida geométrica del evento (como área, longitud o volumen) en relación con la medida del espacio total. Esto se expresa como:

$$
P(E) = \frac{\text{Medida del evento}}{\text{Medida del espacio total}}
$$

**Uniformidad**

El enfoque geométrico asume que los puntos dentro del espacio total están distribuidos de manera uniforme. Esto significa que cada punto tiene la misma probabilidad de ser seleccionado.

**Axiomas de Kolmogórov**

Aunque no son exclusivos del enfoque geométrico, los axiomas de Kolmogórov también aplican aquí, garantizando que las probabilidades sean valores entre 0 y 1, y que la suma de las probabilidades de eventos mutuamente excluyentes sea igual a la probabilidad de su unión.

**Enunciado del Teorema Fundamental de la Probabilidad Geométrica**

Sea 𝑆 un espacio muestral con medida finita 𝜇(𝑆)>0, y 𝐴 un subconjunto medible de 𝑆. Entonces, la probabilidad geométrica de 𝐴, definida como:

$$P(A) = \frac{𝜇(A)}{𝜇(S)},$$

está bien definida y satisface los axiomas fundamentales de la probabilidad.

**Aplicaciones del Teorema Fundamental**

**1. Lanzamiento de dardos:** Determinar la probabilidad de que caiga en una región específica (un círculo, triángulo, etc.).

**2. Selección al azar:** Calcular la probabilidad de que un punto cumpla con ciertas condiciones en un espacio continuo.

**3. Problemas de ingeniería:** Evaluar riesgos en regiones específicas de diseño.

### Problema del encuentro
---
Dos personas acuerdan encontrarse entre las 0 y 60 minutos. Cada una espera 15 minutos. ¿Cuál es la probabilidad de que se encuentren?

Definimos:

- $x$ : Hora de llegada de la persona A
- $y$ : Hora de llegada de la persona B
- Ω : Cuadrado de $[0.6]$ x $[0.6]$
- Evento favorable : $|x-y| \leq 15$

Desarrollo:

**Área total** = $[0.6]$ x $[0.6]$ = $3600$

**Área del evento** = $3600-2 * \frac{1}{2} * 45^2 = 1575$

$$P(encuentro) = \frac{1575}{3600} = \frac{7}{16}$$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 60, 500)
y1 = x - 15
y2 = x + 15

plt.figure(figsize=(8, 8))
plt.fill_between(x, y1, y2, where=(y1 >= 0) & (y2 <= 60), color='skyblue', alpha=0.6, label='Zona de encuentro')
plt.plot(x, y1, 'r--', label='y = x - 15')
plt.plot(x, y2, 'g--', label='y = x + 15')
plt.xlim(0, 60)
plt.ylim(0, 60)
plt.xlabel('Hora de llegada de A (min)')
plt.ylabel('Hora de llegada de B (min)')
plt.title('Problema del Encuentro')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()
```

### Lanzamiento de un dardo en un tablero cuadrado
---

Un tablero cuadrado tiene un lado de 10 cm, y en su interior hay un círculo de radio 3 cm centrado en el tablero. Si lanzas un dardo al tablero, ¿cuál es la probabilidad de que caiga dentro del círculo?

**Espacio total:** El área del tablero cuadrado es:

$$A_{\text{cuadrado}} = 10 \times 10 = 100 \, {cm}^2$$

**Evento:** El área del círculo es:

$$A_{\text{circulo}} = π(3)^2 = 9π cm^2$$

**Probabilidad geométrica:**

$$P(E) = \frac{A_{\text{circulo}}}{A_{\text{cuadrado}}} = \frac{9π}{100} ≈ 0.283$$

En Python sería de la siguiente forma:


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del cuadrado y círculo
lado_cuadrado = 10  # cm
radio_circulo = 3  # cm

# Crear coordenadas para el círculo
theta = np.linspace(0, 2 * np.pi, 500)  # Ángulo en radianes
x_circulo = radio_circulo * np.cos(theta)  # Coordenadas X del círculo
y_circulo = radio_circulo * np.sin(theta)  # Coordenadas Y del círculo

# Coordenadas del cuadrado
x_cuadrado = [-lado_cuadrado/2, lado_cuadrado/2, lado_cuadrado/2, -lado_cuadrado/2, -lado_cuadrado/2]
y_cuadrado = [-lado_cuadrado/2, -lado_cuadrado/2, lado_cuadrado/2, lado_cuadrado/2, -lado_cuadrado/2]

# Crear gráfica
plt.figure(figsize=(8, 8))
plt.fill(x_cuadrado, y_cuadrado, color='lightblue', alpha=0.5, label='Espacio Total (Cuadrado)')
plt.fill(x_circulo, y_circulo, color='pink', alpha=0.7, label='Evento (Círculo)')
plt.plot(x_circulo, y_circulo, color='red', linewidth=1.5, label='Límite del Círculo')
plt.plot(x_cuadrado, y_cuadrado, color='blue', linewidth=1.5, label='Límite del Cuadrado')

# Configurar límites y aspecto
plt.xlim(-lado_cuadrado/2 - 1, lado_cuadrado/2 + 1)
plt.ylim(-lado_cuadrado/2 - 1, lado_cuadrado/2 + 1)
plt.gca().set_aspect('equal')  # Mantener aspecto cuadrado

# Etiquetas y leyenda
plt.title('Probabilidad Geométrica: Círculo Inscrito en un Cuadrado')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.legend()
plt.grid(True)

# Mostrar gráfica
plt.show()
```

# Enfoque frecuentista
---

El enfoque frecuentista interpreta la probabilidad como la frecuencia relativa de la ocurrencia de un evento en un número suficientemente grande de repeticiones de un experimento. En otras palabras, la probabilidad de un evento se define como el límite de su frecuencia relativa cuando el número de ensayos tiende a infinito.

Si un experimento se repite 𝑛 veces y el evento 𝐴 ocurre 𝑓(𝐴) veces, la probabilidad de 𝐴 está dada por:

$$
P(A) = \lim_{n \to \infty} \frac{f(A)}{n}
$$

### **Enunciado del Teorema: Ley de los Grandes Números**
---

Sea $X_1, X_2, \dots, X_n$ una secuencia de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) con esperanza matemática $\mathbb{E}[X] = \mu \text{ y varianza } \text{Var}(X) = \sigma^2 < \infty.$
Entonces, para cualquier $\epsilon > 0:$

$$
P\left(\left| \frac{1}{n} \sum_{i=1}^n X_i - \mu \right| \geq \epsilon \right) \to 0 \quad \text{cuando} \quad n \to \infty
$$

Esto implica que la media muestral $\frac{1}{n} \sum_{i=1}^n X_i \xrightarrow{P} \mu$ converge en probabilidad a la media poblacional 𝜇.

Este teorema fundamenta el enfoque frecuentista, ya que garantiza que, al realizar un número suficientemente grande de repeticiones de un experimento, las frecuencias relativas convergen al valor verdadero de la probabilidad del evento. En términos prácticos:

Más repeticiones → mayor precisión: Si lanzamos una moneda miles de veces, la frecuencia relativa de "cara" se aproximará a 𝑃(cara) = 0.5 (si la moneda es justa).

### Ejemplo del volado
---

Vamos a simular $n = 10,000$ lanzamientos de una moneda justa y observar cómo se estabiliza la frecuencia relativa de “sol”.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 10000
resultados = np.random.choice(['sol', 'aguila'], size=n)
frecuencia_cara = np.cumsum(resultados == 'sol') / np.arange(1, n + 1)

plt.figure(figsize=(10, 5))
plt.plot(frecuencia_cara, label='Frecuencia relativa de "sol"', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='Valor teórico: 0.5')
plt.title('Probabilidad Frecuentista: Convergencia de Frecuencia Relativa')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.legend()
plt.show()
```

### **Ejemplo: Juegos de azar**
---

En un casino, se registra que en 𝑛 = 10000 giros de una ruleta, el número **7** aparece 𝑓(7) = 520 veces.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
n_giros = 10000  # Total de giros de la ruleta
resultados = np.random.randint(0, 37, size=n_giros)  # Simulación de números en la ruleta (0 a 36)
frecuencia_relativa_7 = np.cumsum(resultados == 7) / np.arange(1, n_giros + 1)  # Frecuencia acumulada

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(frecuencia_relativa_7, label='Frecuencia relativa de "7"', color='blue')
plt.axhline(y=1/37, color='red', linestyle='--', label='Probabilidad teórica (1/37)')
plt.xlabel('Número de giros de la ruleta')
plt.ylabel('Frecuencia relativa')
plt.title('Convergencia Frecuentista: Número "7" en la Ruleta')
plt.legend()
plt.grid()
plt.show()
```

### Ejemplo Encuestas de opinión
---

En una encuesta realizada a 𝑛 = 500 personas, 𝑓(𝐴) = 200 personas respondieron que prefieren el producto A.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
n_encuestas = 1000  # Total de encuestados
preferencia_A_prob = 0.4  # Probabilidad teórica de preferir el producto A
encuestas = np.random.choice(["A", "Otros"], size=n_encuestas, p=[preferencia_A_prob, 1 - preferencia_A_prob])  # Simulación
frecuencia_relativa_A = np.cumsum(encuestas == "A") / np.arange(1, n_encuestas + 1)  # Frecuencia acumulada

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(frecuencia_relativa_A, label='Frecuencia relativa de preferencia por A', color='blue')
plt.axhline(y=preferencia_A_prob, color='red', linestyle='--', label='Probabilidad teórica (0.4)')
plt.xlabel('Número de encuestados')
plt.ylabel('Frecuencia relativa')
plt.title('Convergencia Frecuentista: Preferencia por el producto A')
plt.legend()
plt.grid()
plt.show()
```
