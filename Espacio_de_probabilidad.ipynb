# Enfoque cl치sico
---

Este es el m치s intuitivo y se usa cuando todos los resultados posibles son igualmente probables. Por ejemplo, al lanzar un dado, cada cara tiene la misma probabilidad de aparecer. Aqu칤 se calcula la probabilidad como:

$$
P(A) = \frac{\text{N칰mero de casos favorables}}{\text{N칰mero total de casos posibles}}
$$

**Teorema (Enfoque Cl치sico de la Probabilidad):**

Para cualquier evento 洧냢 en un espacio muestral finito 洧녡 con resultados equiprobables, la probabilidad de 洧냢 est치 dada por:

$$
P(A) = \frac{|A|}{|S|}
$$

donde:

$|A|$ es el n칰mero de casos favorables al evento 洧냢

$|S|$ es el n칰mero total de resultados en el espacio muestral 洧녡

El enfoque cl치sico de la probabilidad tiene fundamentos s칩lidos en la teor칤a matem치tica y l칩gica. Se basa en los siguientes principios y conceptos:

**Principio de equiprobabilidad**

El enfoque cl치sico asume que todos los resultados del espacio muestral son igualmente probables. Este principio es v치lido en situaciones donde no hay razones para pensar que un resultado sea m치s probable que otro (por ejemplo, en el lanzamiento de un dado justo).

**Definici칩n axiom치tica**

El enfoque cl치sico es coherente con los axiomas de Kolmog칩rov, que son la base matem치tica de la teor칤a de probabilidades. Estos axiomas garantizan que la probabilidad sea un valor entre 0 y 1, y que la suma de las probabilidades de todos los resultados en el espacio muestral sea igual a 1.

**Aplicaciones pr치cticas**

Este enfoque se aplica principalmente en problemas de juegos de azar y situaciones idealizadas donde las condiciones son controladas y los resultados son discretos y finitos.

**Relaci칩n con el concepto de proporci칩n**

La f칩rmula cl치sica $P(A) = \frac{casos맍avorables}{casos맗osibles}$ utiliza el concepto de proporci칩n, que es fundamental en matem치ticas y l칩gica.

Para este enfoque vamos a revisar algunos conceptos clave, como por ejemplo:

**Espacio muestral (풜)**

Es el conjunto que incluye todos los posibles resultados de un experimento aleatorio. Por ejemplo, si lanzamos un dado, el espacio muestral ser칤a:

$$
풜 = \{1, 2, 3, 4, 5, 6\}
$$

**Eventos**

Un evento es un subconjunto del espacio muestral. Representa una situaci칩n espec칤fica o un conjunto de resultados que satisfacen una condici칩n dada. Por ejemplo, el evento "n칰mero par" al lanzar un dado ser칤a:

Aqu칤 tenemos ejemplos de estos dos conceptos:

```python
# Definir el espacio muestral (lanzamiento de un dado)
espacio_muestral = [1, 2, 3, 4, 5, 6]

# Definir un evento: n칰meros pares
evento_numeros_pares = [x for x in espacio_muestral if x % 2 == 0]

print("Espacio Muestral:", espacio_muestral)
print("Evento - N칰meros pares:", evento_numeros_pares)
```

**Probabilidad condicional**

Es la probabilidad de que ocurra un evento 洧냢 dado que otro evento 洧냣 ya ha ocurrido. Se calcula mediante la f칩rmula:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)},
$$

donde $P(A \mid B)$ es la probabilidad de que ocurran ambos eventos simult치neamente.

```python
# Supongamos que tenemos un espacio muestral con colores y formas
espacio_muestral = [('rojo', 'c칤rculo'), ('azul', 'c칤rculo'), ('rojo', 'cuadrado'), ('azul', 'cuadrado')]

# Calcular la probabilidad condicional de que sea rojo dado que es un c칤rculo
evento_circulo = [x for x in espacio_muestral if x[1] == 'c칤rculo']
evento_rojo_y_circulo = [x for x in evento_circulo if x[0] == 'rojo']

probabilidad_condicional = len(evento_rojo_y_circulo) / len(evento_circulo)

print("Probabilidad de que sea rojo dado que es c칤rculo:", probabilidad_condicional)
```

**Independencia**

Dos eventos 洧냢 y 洧냣 son independientes si la ocurrencia de uno no afecta la probabilidad del otro. Matem치ticamente, esto se expresa como:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

```python
# Definir eventos independientes: lanzar dos dados
import random

def lanzar_dado():
    return random.randint(1, 6)

# Simulaci칩n de lanzamientos
lanzamientos = [(lanzar_dado(), lanzar_dado()) for _ in range(1000)]

# Calcular si los eventos son independientes
evento_suma_7 = [par for par in lanzamientos if sum(par) == 7]
probabilidad_suma_7 = len(evento_suma_7) / len(lanzamientos)

evento_1_en_primer_dado = [par for par in lanzamientos if par[0] == 1]
probabilidad_1_primer_dado = len(evento_1_en_primer_dado) / len(lanzamientos)

print("Probabilidad de obtener suma 7:", probabilidad_suma_7)
print("Probabilidad de obtener un '1' en el primer dado:", probabilidad_1_primer_dado)-
```

# Enfoque geom칠trico
---

Este enfoque es 칰til cuando los resultados posibles de un experimento est치n representados en un espacio continuo. La probabilidad se calcula como una relaci칩n entre las medidas geom칠tricas.

La probabilidad de un evento 洧냦 en el enfoque geom칠trico se define como:

$$
P(E) = \frac{\text{Medida del evento (치rea, longitud, volumen, etc.)}}{\text{Medida del espacio total}}
$$

**Espacio total:** Representa el espacio continuo donde puede ocurrir el experimento (por ejemplo, un segmento, un plano o un volumen).

**Evento:** Es un subconjunto del espacio total que cumple ciertas condiciones (por ejemplo, un intervalo dentro de un segmento, un c칤rculo dentro de un cuadrado, etc.).

Este enfoque es ideal para problemas donde los resultados no son discretos, sino que est치n distribuidos de manera continua.

Este enfoque se basa en principios matem치ticos s칩lidos y en conceptos fundamentales de la teor칤a de medidas. Estos principios incluyen:

**Proporcionalidad de la medida**

La probabilidad de un evento es proporcional a la medida geom칠trica del evento (como 치rea, longitud o volumen) en relaci칩n con la medida del espacio total. Esto se expresa como:

$$
P(E) = \frac{\text{Medida맋el만vento}}{\text{Medida맋el만spacio맚otal}}
$$

**Uniformidad**

El enfoque geom칠trico asume que los puntos dentro del espacio total est치n distribuidos de manera uniforme. Esto significa que cada punto tiene la misma probabilidad de ser seleccionado.

**Axiomas de Kolmog칩rov**

Aunque no son exclusivos del enfoque geom칠trico, los axiomas de Kolmog칩rov tambi칠n aplican aqu칤, garantizando que las probabilidades sean valores entre 0 y 1, y que la suma de las probabilidades de eventos mutuamente excluyentes sea igual a la probabilidad de su uni칩n.

**Enunciado del Teorema Fundamental de la Probabilidad Geom칠trica**

Sea 洧녡 un espacio muestral con medida finita 洧랞(洧녡)>0, y 洧냢 un subconjunto medible de 洧녡. Entonces, la probabilidad geom칠trica de 洧냢, definida como:

$$P(A) = \frac{洧랞(A)}{洧랞(S)},$$

est치 bien definida y satisface los axiomas fundamentales de la probabilidad.

**Aplicaciones del Teorema Fundamental**

**1. Lanzamiento de dardos:** Determinar la probabilidad de que caiga en una regi칩n espec칤fica (un c칤rculo, tri치ngulo, etc.).

**2. Selecci칩n al azar:** Calcular la probabilidad de que un punto cumpla con ciertas condiciones en un espacio continuo.

### Problema del encuentro
---
Dos personas acuerdan encontrarse entre las 0 y 60 minutos. Cada una espera 15 minutos. 쮺u치l es la probabilidad de que se encuentren?

Definimos:

- $x$ : Hora de llegada de la persona A
- $y$ : Hora de llegada de la persona B
- 풜 : Cuadrado de $[0.6]$ x $[0.6]$
- Evento favorable : $|x-y| \leq 15$

Desarrollo:

**츼rea total** = $[0.6]$ x $[0.6]$ = $3600$

**츼rea del evento** = $3600-2 * \frac{1}{2} * 45^2 = 1575$

$$P(encuentro) = \frac{1575}{3600} = \frac{7}{16}$$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 60, 500)
y1 = x - 15
y2 = x + 15

plt.figure(figsize=(8, 8))
plt.fill_between(x, y1, y2, where=(y1 >= 0) & (y2 <= 60), color='skyblue', alpha=0.6, label='Zona de encuentro')
plt.plot(x, y1, 'r--', label='y = x - 15')
plt.plot(x, y2, 'g--', label='y = x + 15')
plt.xlim(0, 60)
plt.ylim(0, 60)
plt.xlabel('Hora de llegada de A (min)')
plt.ylabel('Hora de llegada de B (min)')
plt.title('Problema del Encuentro')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()
```

### Lanzamiento de un dardo en un tablero cuadrado
---

Un tablero cuadrado tiene un lado de 10 cm, y en su interior hay un c칤rculo de radio 3 cm centrado en el tablero. Si lanzas un dardo al tablero, 쯖u치l es la probabilidad de que caiga dentro del c칤rculo?

**Espacio total:** El 치rea del tablero cuadrado es:

$$A_{\text{cuadrado}} = 10 \times 10 = 100 \, {cm}^2$$

**Evento:** El 치rea del c칤rculo es:

$$A_{\text{circulo}} = (3)^2 = 9 cm^2$$

**Probabilidad geom칠trica:**

$$P(E) = \frac{A_{\text{circulo}}}{A_{\text{cuadrado}}} = \frac{9}{100} 곋 0.283$$

En Python ser칤a de la siguiente forma:


```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metros del cuadrado y c칤rculo
lado_cuadrado = 10  # cm
radio_circulo = 3  # cm

# Crear coordenadas para el c칤rculo
theta = np.linspace(0, 2 * np.pi, 500)  # 츼ngulo en radianes
x_circulo = radio_circulo * np.cos(theta)  # Coordenadas X del c칤rculo
y_circulo = radio_circulo * np.sin(theta)  # Coordenadas Y del c칤rculo

# Coordenadas del cuadrado
x_cuadrado = [-lado_cuadrado/2, lado_cuadrado/2, lado_cuadrado/2, -lado_cuadrado/2, -lado_cuadrado/2]
y_cuadrado = [-lado_cuadrado/2, -lado_cuadrado/2, lado_cuadrado/2, lado_cuadrado/2, -lado_cuadrado/2]

# Crear gr치fica
plt.figure(figsize=(8, 8))
plt.fill(x_cuadrado, y_cuadrado, color='lightblue', alpha=0.5, label='Espacio Total (Cuadrado)')
plt.fill(x_circulo, y_circulo, color='pink', alpha=0.7, label='Evento (C칤rculo)')
plt.plot(x_circulo, y_circulo, color='red', linewidth=1.5, label='L칤mite del C칤rculo')
plt.plot(x_cuadrado, y_cuadrado, color='blue', linewidth=1.5, label='L칤mite del Cuadrado')

# Configurar l칤mites y aspecto
plt.xlim(-lado_cuadrado/2 - 1, lado_cuadrado/2 + 1)
plt.ylim(-lado_cuadrado/2 - 1, lado_cuadrado/2 + 1)
plt.gca().set_aspect('equal')  # Mantener aspecto cuadrado

# Etiquetas y leyenda
plt.title('Probabilidad Geom칠trica: C칤rculo Inscrito en un Cuadrado')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.legend()
plt.grid(True)

# Mostrar gr치fica
plt.show()
```

# Enfoque frecuentista
---

El enfoque frecuentista interpreta la probabilidad como la frecuencia relativa de la ocurrencia de un evento en un n칰mero suficientemente grande de repeticiones de un experimento. En otras palabras, la probabilidad de un evento se define como el l칤mite de su frecuencia relativa cuando el n칰mero de ensayos tiende a infinito.

Si un experimento se repite 洧녵 veces y el evento 洧냢 ocurre 洧녭(洧냢) veces, la probabilidad de 洧냢 est치 dada por:

$$
P(A) = \lim_{n \to \infty} \frac{f(A)}{n}
$$

### **Enunciado del Teorema: Ley de los Grandes N칰meros**
---

Sea $X_1, X_2, \dots, X_n$ una secuencia de variables aleatorias independientes e id칠nticamente distribuidas (i.i.d.) con esperanza matem치tica $\mathbb{E}[X] = \mu \text{ y varianza } \text{Var}(X) = \sigma^2 < \infty.$
Entonces, para cualquier $\epsilon > 0:$

$$
P\left(\left| \frac{1}{n} \sum_{i=1}^n X_i - \mu \right| \geq \epsilon \right) \to 0 \quad \text{cuando} \quad n \to \infty
$$

Esto implica que la media muestral $\frac{1}{n} \sum_{i=1}^n X_i \xrightarrow{P} \mu$ converge en probabilidad a la media poblacional 洧랞.

Este teorema fundamenta el enfoque frecuentista, ya que garantiza que, al realizar un n칰mero suficientemente grande de repeticiones de un experimento, las frecuencias relativas convergen al valor verdadero de la probabilidad del evento. En t칠rminos pr치cticos:

M치s repeticiones  mayor precisi칩n: Si lanzamos una moneda miles de veces, la frecuencia relativa de "cara" se aproximar치 a 洧녞(cara) = 0.5 (si la moneda es justa).

### Ejemplo del volado
---

Vamos a simular $n = 10,000$ lanzamientos de una moneda justa y observar c칩mo se estabiliza la frecuencia relativa de 랍ol.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 10000
resultados = np.random.choice(['sol', 'aguila'], size=n)
frecuencia_cara = np.cumsum(resultados == 'sol') / np.arange(1, n + 1)

plt.figure(figsize=(10, 5))
plt.plot(frecuencia_cara, label='Frecuencia relativa de "sol"', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='Valor te칩rico: 0.5')
plt.title('Probabilidad Frecuentista: Convergencia de Frecuencia Relativa')
plt.xlabel('N칰mero de lanzamientos')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.legend()
plt.show()
```
